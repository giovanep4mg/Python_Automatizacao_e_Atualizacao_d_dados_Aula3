#### Desafio: 

Trabalhamos em uma importadora e o preço dos nossos produtos é vinculado a cotação de:
- Dólar
- Euro
- Ouro

Precisamos pegar na internet, de forma automática, a cotação desses 3 itens e saber quanto devemos cobrar pelos nossos produtos, considerando uma margem de contribuição que temos na nossa base de dados.

Base de Dados: https://drive.google.com/drive/folders/1KmAdo593nD8J9QBaZxPOG1yxHZua4Rtv?usp=sharing

Para isso, vamos criar uma automação web:

```python
# instalar selenium 

from selenium import webdriver #>>>permite criar um navegador 
from selenium.webdriver.common.keys import Keys  # >>> permite vc escrever no navegador 
from selenium.webdriver.common.by import By #>>> permite voce selecionar itens do navegador 


```

```python
navegador = webdriver.Chrome()
#abri uma nova janela do navegador 

# abri navegador
navegador.get('https://www.google.com.br')

#pesquizar cotacao
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('cotacao dolar')
#encontra o local de pesquisa do google, e digita o que eu quero pesquisar

# pesquisar cotacao do dolar 
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER )
#local de pesquisa,escrever o que eu quero ,e clicar em enter                                                                                                       

#pegar cotacao do dolar 
cotacao_dolar = navegador.find_element('xpath', '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)

#####   pegar a cotacao do euro 
navegador = webdriver.Chrome()
#abri uma nova janela do navegador 

# abri navegador
navegador.get('https://www.google.com.br')

#pesquizar cotacao
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('cotacao euro')
#encontra o local de pesquisa do google, e digita o que eu quero pesquisar

# pesquisar cotacao do dolar 
navegador.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER )
#local de pesquisa,escrever o que eu quero ,e clicar em enter                                                                                                       

#pegar cotacao do dolar 
cotacao_euro  = navegador.find_element('xpath', '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_euro)

#### pegar a cotacao do ouro
navegador.get('https://www.melhorcambio.com/ouro-hoje')
#entra no site

cotacao_ouro = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
#pesquisar onde o ouro esta e pegar os valores 

#para transforma a (,) em ponto (.)
cotacao_ouro = cotacao_ouro.replace(',','.')

print(cotacao_ouro)

navegador.quit()


```

### Agora vamos atualiza a nossa base de preços com as novas cotações

- Importando a base de dados

```python
import pandas as pd 

tabela = pd.read_excel('Produtos.xlsx')

display(tabela)
```

```python
#atualizar coluna de cotacao


#editar coluna de cotacao onde é dolar
tabela.loc[tabela ['Moeda'] == 'Dólar', 'Cotação'] = float (cotacao_dolar) 



tabela.loc[tabela ['Moeda'] == 'Euro', 'Cotação'] = float (cotacao_euro)


tabela.loc[tabela ['Moeda'] == 'Ouro', 'Cotação'] = float (cotacao_ouro)





#atualizar preco de compra = preco orignal * cotacao

tabela['Preço de Compra'] = tabela ['Preço Original'] * tabela['Cotação']


#atualizar preco de venda = preco de compra * margem
tabela['Preço de Venda'] = tabela ['Preço de Compra'] * tabela['Margem']


display(tabela)
```

### Agora vamos exportar a nova base de preços atualizada

```python
tabela.to_excel('Produtos.nova.xlsx', index= False )

# nao salve com o mesmo nome para nao substituir a tabela antiga !
# esse index e para apagar essa linda com numero no comeco da planilha !
```
